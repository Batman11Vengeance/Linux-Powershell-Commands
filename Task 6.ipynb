```python
# Car Price Prediction using Linear Regression

import pandas as pd
from sklearn.linear_model import LinearRegression

# Load dataset
df = pd.read_csv("carprices.csv")

print("Original Dataset:\n")
print(df)

# Create dummy variables for Car Model
dummies = pd.get_dummies(df['Car Model'])

print("\nDummy Variables:\n")
print(dummies)

# Merge original data with dummies
merged = pd.concat([df, dummies], axis='columns')

print("\nMerged Dataset:\n")
print(merged)

# Remove original car model column and one dummy column
# (avoids dummy variable trap)
final = merged.drop(
    ["Car Model", "Mercedez Benz C class"],
    axis='columns'
)

print("\nFinal Dataset:\n")
print(final)

# Separate features and target
X = final.drop('Sell Price($)', axis='columns')
y = final['Sell Price($)']

# Create Linear Regression model
model = LinearRegression()

# Train model
model.fit(X, y)

# Display model accuracy
print("\nAccuracy:", model.score(X, y))

# Predict Mercedes Benz price
mercedes = model.predict([[45000, 4, 0, 0]])

print("\nPredicted price of Mercedes Benz (4 years old, 45000 mileage):")
print(mercedes[0])

# Predict BMW X5 price
bmw = model.predict([[86000, 7, 0, 1]])

print("\nPredicted price of BMW X5 (7 years old, 86000 mileage):")
print(bmw[0])
```
